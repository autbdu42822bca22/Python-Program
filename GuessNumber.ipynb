```python
print("Welcome to Number Guessed Game!!!")
print("Guess the Number Between 1 to 100")
import random
secret_number = random.randint(1,100)
for i in range(1,8):
    try:
        user_input = int(input("enter your guess:"))
        if user_input < 1 or user_input > 100:
          print("Invalied Number! Please enter the Number between 1-100")
        elif user_input > secret_number:
          print("Your Number Higher!")
        elif user_input < secret_number:
          print("Your Number Lesser!")
        else:
          print("Your Guessed Number is Correct")
          break
    except ValueError:
        print("Invaild Character ")
            
if secret_number == user_input:
    print(f"congradulations, you have guesses the number {secret_number} in {i} attemts")
else:
    print("Sorry! All your trials is exceeded.try Again!!")   
```
